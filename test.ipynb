```python
from glob import glob
import re 
import os
```

```python
ptcorpus = os.path.join('resources', 'corpus', 'ptwiki.txt')
ptcorpus2 = os.path.join('resources', 'corpus', 'ptwiki2.txt')


def read_file(path):
    with open(path, 'r', encoding='utf-8') as f:
        data = f.read()

    return data

def write_file(fname, data):
    with open(fname, 'w', encoding='utf-8', errors='ignore') as f:
        f.write(data)
```

```python
text = read_file(ptcorpus2)
pattern = r"[\uf000-\ufaff]"
cleaned_text = re.sub(pattern, "", text)
cleaned_text = cleaned_text.replace('\x0c', '').replace('\ufb02', '')
```

```python
cleaned_text
```

```python
write_file(ptcorpus2, cleaned_text)
```
